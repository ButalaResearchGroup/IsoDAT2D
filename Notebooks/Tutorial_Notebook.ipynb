```python
import ThinFilmDataCreation as tfdc
import numpy as np
import matplotlib.pyplot as plt
```

# Generating the Thin Film Image

```python
#Creating the single-crystal diffraction pattern
spots = tfdc.create_anisotropic(10, 25, 300, size = 200, cmap = 'magma')
spots = spots*15
rotated_spots = tfdc.rotate_image(spots, 45, cmap = 'magma')
```

```python
#Creating the thin-film diffraction pattern from a calibrant material, selecting "18" Ni as the calibrant
pattern = tfdc.create_isotropic(.5, .5e-10, cmap = "magma" )
pattern = pattern*2
```

```python
#Combining the images of the thin-film signal and the single-crystal signal
thin_film = tfdc.combine_image(pattern, rotated_spots, cmap = "magma")
```

## Integrating the created thin film signal

```python
#creating the mask for the integreations

mask = tfdc.create_mask(thin_film, 10)
```

```python
#Integrating for one angle

one_line_integration = tfdc.integrate_image(thin_film, .5, .5e-10, mask=mask, show = True)
```

```python
#rotating and integrating the thin-film mask

data_to_use = tfdc.rotate_and_integrate(thin_film, 10, .5, .5e-10, resolution = 1000, mask = mask)
```
